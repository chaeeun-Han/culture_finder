```python
import numpy as np
import chromedriver_autoinstaller
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import os
from bs4 import BeautifulSoup
import time as t
```

```python
chromedriver_autoinstaller.install()
options = webdriver.ChromeOptions() 
# options.add_argument("--headless")
driver = webdriver.Chrome(options=options)

```

```python
def get_name_address(name,retry=False):
    if not retry:
        url = f"https://map.kakao.com/link/search/{name}"
        driver.get(url)
        html = driver.page_source
        soup = BeautifulSoup(html,'lxml')
        t.sleep(2)
        table = soup.find_all("ul", {"class": "placelist"})[0]
        if not table.text:
            return None
    else:
        t.sleep(5)
        html = driver.page_source
        soup = BeautifulSoup(html,'lxml')
        table = soup.find_all("ul", {"class": "placelist"})[0]
        if not table.text:
            return None
    name = table.find_all("a", {"class": "link_name"})[0]
    address = table.find_all("div", {"class": "addr"})[0].find_all("p", {"data-id":"address"})[0]
    contact = table.find_all("div", {"class": "contact clickArea"})[0]
    phone = contact.find_all("span", {"data-id": "phone"})[0].text
    homepage = contact.find_all("a", {"class": "homepage"})[0]
    return name['title'], address['title'], phone, homepage['href']
```

```python
get_name_address("고랴대학교")
```

```python
df = pd.read_csv("origc.csv")
```

```python
df.drop(["Unnamed: 4"], axis=1, inplace=True)
df['시설명'] = df['시설명'].str.replace('\n', ' ')
new = []
df
```
